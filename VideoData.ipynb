```python
import os
import numpy as np
import cv2
import matplotlib.pyplot as plt
from PIL import Image
import pandas as pd
from datetime import datetime
```

#### Video Data

```python
image_path = '2024-03-12/'
max_images = 20  # Maximum number of images to load
count = 0  # Counter for loaded images

# Define the coordinates of the region to be cropped (x, y, width, height)
x, y, w, h = 200, 200, 1600, 1600
# Create a list to store cropped images
cropped_images_list = []

# List to store images
images = []
epochs = []
timestamp = []
images2 = []
files = os.listdir(image_path)
print(pd.DataFrame(files))
print(len(files))
# Iterate over files in the folder
for filename in os.listdir(image_path):
    # Check if the maximum number of images has been reached
    if count >= max_images:
        break
    
    # Construct the full path t the image
    image_file_path = os.path.join(image_path, filename)
    
    #Get epoch from filename
    epoch = filename[0:len(filename)-4]
    epochs.append(epoch)

    #Convert to timestamp
    timestamp = datetime.fromtimestamp(int(epoch)).strftime('%Y-%m-%d %H:%M:%S')
    # Load the image using OpenCV
    img = cv2.imread(image_file_path)
    img2 = Image.open(image_file_path)
    images.append(img)
    images2.append(img2)
    
    # Increment the counter
    count += 1
```

### Create timestamps

```python
df = pd.DataFrame({'FileName': epochs, 'Image': images, 'Image2': images2, 'Timestamp':timestamp})
df
```
